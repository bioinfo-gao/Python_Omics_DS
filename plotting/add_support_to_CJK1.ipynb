```python
import matplotlib.pyplot as plt
from matplotlib import font_manager

# Replace 'Noto Sans CJK JP' with the name of a CJK-supported font on your system.
# For example, on Windows, you might use 'Microsoft YaHei'.
# Chinese Japanese and Korean fonts are supported by default in matplotlib.
font_properties = font_manager.FontProperties(family='Noto Sans CJK CN')

plt.title("测试", fontproperties=font_properties)
plt.xlabel("测试", fontproperties=font_properties)
plt.ylabel("测试", fontproperties=font_properties)
plt.plot([1, 2, 3])
plt.show()
```
