```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 设置风格
#sns.set(style="whitegrid") # 另一种风格 white, dark, whitegrid, darkgrid, ticks 
sns.set_theme(style="darkgrid") # 另一种风格 white, dark, whitegrid, darkgrid, ticks 

# 生成数据
x  = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建图表
plt.figure(figsize=(10, 2))
plt.plot(x, y1, label="Sine Wave", color="b", linewidth=2)
plt.plot(x, y2, label="Cosine Wave", color="r", linestyle="--", linewidth=2)

# 添加装饰
plt.fill_between(x, y1, y2, color="gray", alpha=0.35)
plt.title("Line Plot", fontsize=15)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.legend()
plt.show()
```

2. 散点图

```python
import matplotlib.pyplot as plt
import seaborn           as sns
import numpy             as np
# 设置风格
sns.set_theme(style="whitegrid")

#生成数据
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100) # for colormap cmap='viridis' , dark to light

# The colormap used here is 'viridis', as specified by the argument cmap='viridis' in plt.scatter().

# colors数组包含100个0到1之间的随机数，这些数值将用于映射散点图中每个点的颜色。
# 实际应用：在散点图中，colors数组的每个数值会被映射为一种颜色，数值越大颜色越偏向colormap的高端（如viridis的黄色），数值越小颜色越偏向低端（如viridis的紫色）。
#print(colors)

sizes = 1000 * np.random.rand(100)
# 创建图表
plt.figure(figsize=(10, 6))
#plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='cool')
plt.colorbar()
# 添加装饰
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.show()
```

3. Bar图

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 设置风格
sns.set_theme(style="whitegrid")

# 生成数据
categories = ['A', 'B', 'C', 'D']
values1 = [5, 7, 8, 6]
values2 = [3, 4, 5, 2]

# 创建图表
fig, ax = plt.subplots (figsize=(10, 6))
bar1 = ax.bar(categories, values1, label='Group 1')
bar2 = ax.bar(categories, values2, bottom=values1, label='Group 2')

# 添加装饰
ax.set_title("Stacked Bar Chart", fontsize=15) #<<============ title ===============
ax.set_xlabel("Categories", fontsize=12)
ax.set_ylabel("Values", fontsize=12)
ax.legend()


# 添加数值标签
for rect in bar1 + bar2: 
    height = rect.get_height() 
    print(height)
    ax.annotate(f'(height)', 
                xy=(rect.get_x() + rect.get_width() / 2, height), #  xy is the point to label at 
                xytext=(0,30), #  xytext=(0,50), #  xytext=(0,3), 
                textcoords= "offset points",    # how to position the text offset points by (x,y) from the given xy
                ha='center', # 'center', 'right', 'left' meaning horizontalalignment
                va='bottom') # 'center', 'top', 'bottom' meaning verticalalignment
    print(rect.get_x() , rect.get_width() , height, "offset points", 'center', 'bottom')
plt.show()
```

6 蜘蛛图

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置风格
sns.set_theme(style="whitegrid")

# 数据准备
labels = np.array (['A', 'B', 'C', 'D', 'E'])
stats = [10, 20, 30, 40, 50]
stats2 = [30, 10, 20, 30, 40]

# 创建蜘蛛图
angles = np.linspace(0, 2* np.pi, len (labels), endpoint=False).tolist() # 2 pi 角度划分
#print(angles)

stats   = np.concatenate((stats,  [stats[0]] )) # original list + first element
stats2  = np.concatenate((stats2, [stats2[0]])) # original list + first element

angles += angles[:1] # original list + first element
# print(stats)
# print(stats2)
# print(angles)

fig, ax = plt.subplots (figsize=(3, 3), subplot_kw=dict(polar=True))

ax.fill(angles, stats,  color='blue', alpha=0.25, label='Group 1') # inner area of the first group
ax.plot(angles, stats,  color='blue', linewidth = 2) # 第一组的边框线
ax.fill(angles, stats2, color='red',  alpha=0.5, label='Group 2') # inner area of the second group
ax.plot(angles, stats2, color='red',  linewidth=2) # 第二组的边框线

ax.set_yticklabels([])
ax.set_xticks (angles[:-1]) # 去掉最后一个角度标签，因为它与第一个标签重合
#ax.set_rlabel_position(0) # 去掉右侧坐标轴标签
ax.set_xticklabels(labels, fontsize=10, color='black')
#ax.grid(False)    # default 显示网格线
#ax.spines['polar'].set_visible(False) # 去掉极坐标的边

# 添加标题和图例
plt.title('Enhanced Spider Chart', size=20, color='black', y = 1.1 )
plt.legend (loc='upper right', bbox_to_anchor=(0.1, 0.1)) # 图例位置 (upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center)
plt.show()
```

7 双轴图

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# 设置风格
sns.set_style(style="whitegrid") # set_style  set_theme same effect 

#生成数据
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = 2*np.cos(x)

# 创建图表
fig, ax1 = plt.subplots (figsize=(10, 6))
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')

# 添加装饰
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Sine', color='g')
ax2.set_ylabel('2 * Cosine', color='b') # <<============ 2 * Cosine ===============
plt.title('Dual Axis Plot')
plt.show()
```

9. 带状图

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# 设置风格
sns.set_style(style="whitegrid")

# 生成数据
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.sin(x) + np.random.normal(0,0.1, 100)

# 创建图表
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave')
#plt.plot(x, z, label='Noisy Sine Wave', alpha=0.7)

plt.fill_between (x, y, z, where=(y >= z), interpolate=True, color='green', alpha=0.3) # where=(y >= z) 条件填充 color='green' y >= z
plt.fill_between (x, y, z, where=(y < z), interpolate=True, color='red', alpha=0.3) # where=(y < z) 条件填充 color='red' y < z

# 添加装饰
plt.title("Band Chart", fontsize=15)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.legend()
plt.show()
```

10. 等高线图

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# 设置风格
sns.set_style(style="white")

# 数据准备
x = np.linspace (-5, 5, 50)
y = np.linspace (-5, 5, 50)

X, Y = np.meshgrid(x, y) # 网格化坐标 

Z=np.sin(np. sqrt(X ** 2 + Y ** 2) ) # 计算Z值 from f(x, y) = sin(sqrt(x^2 + y^2)) 

# 创建等高线图
plt.figure(figsize=(10, 6))
contour = plt.contourf (X, Y, Z, cmap='coolwarm', levels=20)
plt.colorbar (contour)

# 添加装饰
plt.title('Contour Plot', fontsize=15)
plt.show()
```

11. 极坐标图

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# 设置风格
sns.set_style(style="white")

# 数据准备
theta = np.linspace(0, 2*np.pi, 1000)
r     = np.abs(np.sin(3*theta) * np.cos(2*theta))
#r     = np.abs(np.sin(theta) * np.cos(theta))

# 创建极坐标图
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True) # polar=True 极坐标 <<=====================================
ax.plot(theta, r, color='b', linewidth=2)

# 添加装饰
plt.title('Polar Plot', fontsize=15)
plt.show()
```

12. 3D 曲面图

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
# 设置风格
sns.set_style(style="white")
# 数据准备
x = np.linspace (-5, 5, 50)
y = np.linspace (-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z=np.sin(np. sqrt(X ** 2 + Y ** 2) )

# 创建3D曲面图
fig = plt.figure(figsize=(10, 6)) 
ax = fig.add_subplot(111, projection='3d') # projection='3d' 3D图 <<=====================================
surf = ax.plot_surface (X, Y, Z, cmap='coolwarm') #viridis # cool   # plasma # spring #win

# 添加装饰
plt.title('3D Surface Plot', fontsize=15)
plt.show()
```

13. 3D 散点图

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
# 设置风格
sns.set_style(style="white")
# 数据准备
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)

# 创建3D散点图
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter (x, y, z, c=z, cmap='cool') # cmap='viridis' 颜色映射 Viridis, plasma, inferno, magma, cividis prism_r

# 添加装饰
fig.colorbar(scatter)
plt.title('3D Scatter Plot', fontsize=15)
plt.show()
```

18. Facet Grid图 ====== 分组画图：

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 生成数据
tips = sns.load_dataset("tips")
print(tips.head())
print(tips.shape)


# 创建图表
g = sns.FacetGrid(tips, col="time", row="smoker", margin_titles=True) # col="time" 列分组  row="smoker" 行分组 margin_titles=True 显示子图标题
g.map(sns.scatterplot, "total_bill", "tip", alpha=.6) # alpha=.7 透明度
#g.set(xlim=(0, 60), ylim=(0, 12)) # 设置x轴和y轴范围
g.add_legend()

# 添加装饰
plt.suptitle("Facet Grid", y = 1.02, fontsize=15)
plt.show()
```
